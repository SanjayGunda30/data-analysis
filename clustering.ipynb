# Kmean cluster

```python
from sklearn.cluster import KMeans
```

```python
data2 = pd.read_excel('/content/EastWestAirlines.xlsx',sheet_name='data')
```

```python
from sklearn.preprocessing import minmax_scale
```

```python
datascaler = minmax_scale(data2)
data2 = pd.DataFrame(datascaler,columns=data2.columns)
```

```python
data2
```

```python
#now we apply kmean
#to know k value we use for loop
from sklearn.cluster import KMeans
Wss = []
for i in range(1,11):
  Kmean = KMeans(n_clusters=i).fit(data2)
  Wss.append(Kmean.inertia_)
plt.plot(range(1,11),Wss)
plt.title('Elbow crover')
plt.xlabel('kmeans values')
plt.ylabel('wss value')
```

```python
from sklearn.cluster import KMeans
Clf= KMeans(n_clusters=3)
Clf.fit(data2)
```

```python
cluster = Clf.predict(data2)
```

```python
data = pd.DataFrame(cluster,columns=['cluster'])
```

```python
data = pd.concat([data2,data],axis=1)
```

```python
data.head()
```

```python
Clf.cluster_centers_
```

```python
Clf.labels_
```

```python
Clf.inertia_
```

```python
plt.style.use('classic')
print(plt.style.available)
```

```python
data.iloc[:,1:].groupby(data.cluster).mean()
```

```python
data.plot("Balance","Qual_miles",c=Clf.labels_,kind="scatter",s=50,cmap=plt.cm.coolwarm_r)
```

```python
from sklearn.metrics import silhouette_score
print(silhouette_score(data2,Clf.labels_))
```

# DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
```

```python
data = pd.read_excel('/content/EastWestAirlines.xlsx',sheet_name='data')
```

```python
scaler = StandardScaler()
```

```python
datascaler = scaler.fit_transform(data)
```

```python
data = pd.DataFrame(datascaler,columns=data.columns)
```

```python
data
```

```python
# @title Qual_miles

from matplotlib import pyplot as plt
data['Qual_miles'].plot(kind='line', figsize=(8, 4), title='Qual_miles')
plt.gca().spines[['top', 'right']].set_visible(False)
```

```python
# prompt: i want do drop  id# Code

data = data.drop('ID#', axis=1)
```

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
```

```python
from sklearn.neighbors import NearestNeighbors
sd = NearestNeighbors(n_neighbors=5).fit(data)
```

```python
distances, indices = sd.kneighbors(data)
```

```python
distances = np.sort(distances,axis=0)
distances = distances[:,1]
plt.plot(distances)
```

```python
distances
```

```python
indices
```

```python
dbscan = DBSCAN(eps=0.5,min_samples=5)
dbscan.fit(data)
```

```python
dbscan.labels_
```

```python
dataa= pd.DataFrame(dbscan.labels_,columns=['cluster'])
```

```python
data = pd.concat([data,dataa],axis=1)
```

```python
data.head()
```

```python
from sklearn.metrics import silhouette_score
```

```python
print(silhouette_score(data,dbscan.labels_))
```
