#Standardization of Numerical Variables
#â—	Objective: To scale numerical variables for uniformity, improving the datasetâ€™s suitability for analytical models.#
â—	Steps:#
	Explain the concept of standardization (z-score normalization).#
	Standardize the numerical columns using the formula: z=x-mu/sigma#
	
	Show before and after comparisons of the data distributions.


```python
#Explain the concept of standardization (z-score normalization).
# ans : standardization the proess of hight values to standard for analysis  of data  (mean = 0 and standard devieation = 1
# ans : it is proess of converting or comparing of data to standard form to do data anylsis and converting the data to have mean is 0 and 
# standard devieation is 1  z = (X -mu)/s/d
# standardization methode
def standardization(i):
    z = (i - i.mean())/i.std()
    return (z)
```
