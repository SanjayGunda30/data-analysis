# Sales Analysis:

Part 1: Sales and Profitability Analysis

```python
Order_df = pd.read_csv('/content/List_of_Orders_55FFC79CF8.csv')
order_details = pd.read_csv('/content/Order_Details_19795F61CF.csv')
```

```python
print(Order_df.head())
print(order_details.head())
```

```python
marge_df = pd.merge(Order_df, order_details, on='Order ID')
```

```python
total_sales = marge_df.groupby('Category')['Amount'].sum()
```

```python
print(total_sales)
```

```python
categorical_analysis = marge_df.groupby('Category').agg(
    total_sales=('Amount', 'sum'),
    total_profit=('Profit', 'sum'),
    average_profit=('Profit', 'mean')
).reset_index()
```

```python
categorical_analysis
```

```python
top_profomance = categorical_analysis.sort_values(by='total_profit', ascending= False).head(1)
under_performance = categorical_analysis.sort_values(by='total_profit', ascending= True).head(1)
```

```python
print(top_profomance)
print(under_performance)
```

"We have good performance in clothing but underperformance in furniture. To increase our profit, we should focus on increasing sales or work on customer feedback to improve customer satisfaction."


## Part 2: Target Achievement Analysis

```python
data = pd.read_csv('/content/Sales_target_DD2E9B96A0.csv')

# Create a DataFrame
sales_target_df = pd.DataFrame(data)

```

```python
sales_target_df
```

```python
# Calculate the percentage change in target sales month-over-month
sales_target_df['Percentage_Change'] = sales_target_df['Target'].pct_change() * 100

# Display the DataFrame with the percentage change
print(sales_target_df)
```

```python
# Define a threshold for significant fluctuations (e.g., 2% in this example)
threshold = 2

# Identify months with significant fluctuations
significant_fluctuations = sales_target_df[abs(sales_target_df['Percentage_Change']) > threshold]

# Display the months with significant fluctuations
print("Months with Significant Fluctuations:")
print(significant_fluctuations)
```

# Part 3: Regional Performance Insights

```python
marge_df.head()
```

```python
# Calculate order count per state
order_count_per_state = marge_df['State'].value_counts().reset_index()
order_count_per_state.columns = ['State', 'Order_Count']

# Identify the top 5 states with the highest order count
top_5_states = order_count_per_state.head(5)
```

```python
top_5_states
```

```python
# Calculate total sales and average profit for each of the top 5 states
state_analysis = marge_df[marge_df['State'].isin(top_5_states['State'])].groupby('State').agg(
    total_sales=('Amount', 'sum'),
    average_profit=('Profit', 'mean')
).reset_index()

# Merge the order count with the state analysis
state_analysis = pd.merge(state_analysis, top_5_states, on='State')

# Display the result
print("Top 5 States with Highest Order Count:")
print(state_analysis)
```

```python
# Highlight any regional disparities in sales or profitability
total_sales_avg = state_analysis['total_sales'].mean()
average_profit_avg = state_analysis['average_profit'].mean()

disparities = state_analysis[(state_analysis['total_sales'] < total_sales_avg) | (state_analysis['average_profit'] < average_profit_avg)]

# Display regions with disparities
print("Regions with Sales or Profitability Disparities:")
print(disparities)


```
